Time series modelling

```python
import statsmodels
```

By visual inspection, the period looks to be $7$ days, in the sense that the call volume $7$ days apart has noteworthy positive correlation ($0.4$).

```python
plt.figure()
plt.plot(acf(nCalls))
plt.xlabel('lag')
plt.ylabel('autocorrelation')
plt.show()
```

Use Gaussian process to model time series

```python
import numpy as np
```

```python
X = np.array(range(len(nCalls.index)))[:,np.newaxis]
Y_ = np.array(nCalls)
# print(X)
# print(Y)
```

Standardize the data

```python
mY = np.mean(Y_)
sdY = np.std(Y_)

Y = Y_ - mY
Y = Y_/sdY
```

Use exponential sine squared kernel because of periodicity, matern since series is rough, and white kernel to allow for noise.

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ExpSineSquared, Matern, WhiteKernel
```

```python
kernel = 1*Matern() + 1 * ExpSineSquared(periodicity=7, periodicity_bounds="fixed") + WhiteKernel()
# kernel = 1 * Matern() + WhiteKernel()
# kernel = 1*Matern()
# kernel = 1 * RBF() + WhiteKernel()
```

```python
reg = GaussianProcessRegressor(kernel=kernel).fit(X, Y)
```

Inspect fitted kernel

The noise variance is about $0.4$.
The Matern length scale is about $3.5$ days.
The Exponential Sine Squared length scale is shorter than the Matern length scale ($1.5$ days).

```python
reg.kernel_
```

Inspect fit on data and visualize uncertainty

```python
mpred, sepred = reg.predict(X, return_std=True)
plt.figure()
plt.plot(mpred, label='pred', color='red')
plt.plot(Y, label='data', color='blue')
plt.legend()
plt.show()
```

Inspect residuals

The residuals have a longer left tail than right tail. Hypothesis tests for normality would probably reject the hypothesis that the residuals came from a normal distribution.

```python
from scipy.stats import describe
```

```python
res = mpred - Y
print(describe(res))

plt.figure()
plt.hist(res)
plt.show()
```

Forecast into the future with uncertainty bands

```python
Xtest = np.linspace(np.amax(X),np.amax(X) + 100, num = 200)[:,np.newaxis]
Ytestpred, Ytestsd = reg.predict(Xtest, return_std=True)
plt.plot(Xtest.squeeze(), Ytestpred, label='pred', color='black')
plt.fill_between(Xtest.squeeze(), Ytestpred-2*Ytestsd, Ytestpred+2*Ytestsd, alpha=0.7)
plt.legend()
plt.show()
```

Convert predictions in standardized space into original space

```python
plt.figure(figsize=(8,5))

# traininig
mpred, sepred = reg.predict(X, return_std=True)
plt.plot(mpred*sdY + mY, label='pred', color='red')
plt.plot(Y_, label='train data', color='blue')

# forecasting
Xtest = np.linspace(np.amax(X),np.amax(X) + 30, num = 200)[:,np.newaxis]
Ytestpred, Ytestsd = reg.predict(Xtest, return_std=True)
plt.plot(Xtest.squeeze(), Ytestpred*sdY + mY, label='forecast', color='black')
plt.fill_between(Xtest.squeeze(), (Ytestpred-2*Ytestsd)*sdY + mY, (Ytestpred+2*Ytestsd)*sdY + mY, alpha=0.7)

plt.xlabel('day')
plt.ylabel('number of calls')
plt.legend()
plt.show()
```
