```python
def read_data(repo, parse_dates=[]):
    return pd.read_csv(repo, skipinitialspace=True, parse_dates=parse_dates)
```

```python
services_enc = read_data('https://raw.githubusercontent.com/datakind/UWWi_DataDive_Dec22/main/uwwi_datasets/uwwi_dataset_services_encoded.csv',
                    parse_dates=['Service_CreateStamp','Service_EditStamp'])

services_enc[:5]
```

```python
# create separate lists

import ast

day_lst = []
start_hour_lst = []
end_hour_lst = []

for row in services_enc['ServiceHoursofoperation_ModuleHoursofoperation.open']:
    for day in ast.literal_eval(row):
        day_lst.append(day['dayOfWeek'])
        start_hour_lst.append(day['start_hour'])
        end_hour_lst.append(day['end_hour'])
```

```python
# assign arbitrary date, corresponding with day of week

day_dt_lst = []

for day in day_lst:
    if day=='Sunday':
        day_dt_lst.append('11/20/2022')
    elif day=='Monday':
        day_dt_lst.append('11/21/2022')
    elif day=='Tuesday':
        day_dt_lst.append('11/22/2022')
    elif day=='Wednesday':
        day_dt_lst.append('11/23/2022')
    elif day=='Thursday':
        day_dt_lst.append('11/24/2022')
    elif day=='Friday':
        day_dt_lst.append('11/25/2022')
    elif day=='Saturday':
        day_dt_lst.append('11/26/2022')

day_dt1_lst = [pd.to_datetime(day) for day in day_dt_lst]
```

```python
# change start hour to datetime format

start_hour_dt_lst = []

for time in start_hour_lst:
    if time==None:
        start_hour_dt_lst.append(0)
    else:
        start_hour_dt_lst.append(time[0])

start_hour_dt1_lst = [pd.to_timedelta(time, unit='h') for time in start_hour_dt_lst]        
```

```python
# change end hour to datetime format

end_hour_dt_lst = []

for time in end_hour_lst:
    if time==None:
        start_hour_dt_lst.append(0)
    else:
        end_hour_dt_lst.append(time[0])

end_hour_dt1_lst = [pd.to_timedelta(time, unit='h') for time in end_hour_dt_lst]        
```

```python
# one master start datetime

start_dt_lst = []

for idc in range(len(end_hour_dt_lst)):
    start_dt_lst.append(str(day_dt1_lst[idc].date()) + ' ' + str(pd.to_timedelta(start_hour_dt1_lst[idc].seconds//3600, unit='h')))

start_dt1_lst = pd.Series(start_dt_lst).map(lambda x: x[0:10]).values + ' ' + pd.Series(start_dt_lst).map(lambda x: x[-8:])
```

```python
start_dt1_lst.max()
```

```python
# one master end datetime

end_dt_lst = []

for idc in range(len(end_hour_dt_lst)):
    end_dt_lst.append(str(day_dt1_lst[idc].date()) + ' ' + str(pd.to_timedelta(end_hour_dt1_lst[idc].seconds//3600, unit='h')))

end_dt1_lst = pd.Series(end_dt_lst).map(lambda x: x[0:10]).values + ' ' + pd.Series(end_dt_lst).map(lambda x: x[-8:])
```

```python
end_dt1_lst.max()
```

```python
# one master "difference of hours" datetime list, a list of all datetimes between "start" and "end" by hour

difference_dt_lst = []

for idc in range(len(end_dt_lst)):
    difference_dt_lst.append(pd.date_range(pd.to_datetime(start_dt1_lst[idc]), pd.to_datetime(end_dt1_lst[idc]), freq='H'))
```

```python
# groupby hour: generate one dataframe for each datetime hour and concatenate all these dataframes togther for a measure of frequency

date_dfs_lst = []

for diff in difference_dt_lst:
    for date in diff:
        date_dfs_lst.append(pd.DataFrame({'Date':date,'Value':[1]}))
```

```python
# received memory error when concatenating full list of dfs; split / slice list
# part 1

all_dt_1 = pd.concat(date_dfs_lst[:len(date_dfs_lst)//2])
```

```python
# part 2

all_dt_2 = pd.concat(date_dfs_lst[len(date_dfs_lst)//2:])
```

```python
# part 3

all_dt = all_dt_1.append(all_dt_2)
```

```python
# part 4

final_dt = all_dt.groupby(['Date']).agg({'Value':'sum'}).reset_index(drop=False)
```

```python
# part 5

df_pivot = final_dt.pivot_table(columns=final_dt['Date'].dt.day,index=final_dt['Date'].dt.hour)
```

```python
# rename columns and set index to hour

names = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']

df_pivot.columns = names
df_pivot.index.name = 'Hour'
```

```python
# generate heatmap

import plotly.express as px
from plotly.offline import *

def plotly_heatmap(df, width, height, x_axis, y_axis, color, title):

    fig = px.imshow(img=df, aspect='auto',x=df.columns.values, y=df.index.values,
                    width=width, height=height, origin='upper',
                    zmin=df_pivot.values.min(), zmax=df_pivot.values.max(),   
                    labels={'x':x_axis,'y':y_axis},
                    color_continuous_scale=color).update_layout({
        'title':{'text':title,'x':0.5,'font':{'size':20}},'yaxis':
        {'tickmode':'linear'}})
    
    return fig

fig = plotly_heatmap(df_pivot, 900, 600, 'Day of the Week', df_pivot.index.name,'Tealgrn','Open Hours of Service: Day by Hour')
```
